```python
import gradio as gr
from huggingface_hub import login
from google.colab import userdata
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel


hf_token = 'hf_fsSNQRPPciCvyeJbsGUyDbxxxxxxxxxxxxxxx'
login(hf_token)


model_id = "pradeepsengarr/my-pricer-2025-08-02_08.44.25"
# Quantization
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)

# Loading model
tokenizer = AutoTokenizer.from_pretrained(model_id, token=hf_token)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quant_config,
    device_map="auto",
    token=hf_token
)
model = PeftModel.from_pretrained(model, model_id, token=hf_token)

# Prediction function
def predict_price(description):
    if not description.strip():
        return "Please enter a product description!"

    # Suggestion
    prompt = f"How much does this cost to the nearest dollar?\n\n{description}\n\nPrice is $"

    # Generate prediction
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")
    outputs = model.generate(**inputs, max_new_tokens=10, do_sample=False)
    full_output = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # after "Price is $"
    predicted_price = full_output.split("Price is $")[-1].strip()

    # Label the output clearly
    return f"**Predicted Price:** ${predicted_price}"

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# Price Prediction Assistant")
    gr.Markdown("Enter a product description, and I'll predict the price based on your fine-tuned model!")

    input_text = gr.Textbox(label="Product Description", placeholder="E.g., A red bicycle with 21 gears")
    output_text = gr.Textbox(label="Prediction")
    submit_button = gr.Button("Predict Price")

    submit_button.click(predict_price, inputs=input_text, outputs=output_text)

# app
demo.launch(share=True)
```
